```python
from sklearn import cluster, datasets
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans, DBSCAN
```

```python
#Create the sample data

moons_data_x, moons_data_y = datasets.make_moons(n_samples=2000, noise=0.09, random_state=7)
```

```python
#plot the data
plt.scatter(moons_data_x[:,0], moons_data_x[:,1], c=moons_data_y, s=10)
```

## k-Means
cluster the dataset using the k-means algorithm

```python
kmeans = KMeans(n_clusters= 2, #TODO: enter a value 
                init='k-means++', 
                max_iter= 480, #TODO: enter a value 
                n_init= 17, #TODO: enter a value 
                random_state=7)

kmeans_labels = kmeans.fit(moons_data_x)
```

```python
#visualize the results from k-means
plt.scatter(moons_data_x[:,0], moons_data_x[:,1], c=kmeans_labels.labels_, s=10)
```

### Question
- What are your thoughts on the cluster(s)? Are there outliers and was k-means able to detect them?

The clusters make sense given what we know about K Means. The algorithm is splitting the data in a linear fashion so even though there are points that make sense visually to be included in the same cluster, the linearity of K Means prevents it from recognizing these. There does appear to be an outlier but K Means was not able to detect it. Detecting that outlier would probably require a significant number of clusters which would be adverse to trying to catch the two present groups.

## DBSCAN
Cluster the dataset using the DBSCAN algorithm

```python
dbscan = DBSCAN(eps=0.2, #TODO: enter a value 
                min_samples=50 #TODO: enter a value 
               )

dbscan_labels = dbscan.fit(moons_data_x)
```

```python
#visualize the results from DBSCAN
plt.scatter(moons_data_x[:,0], moons_data_x[:,1], c=dbscan_labels.labels_, s=10)
```
